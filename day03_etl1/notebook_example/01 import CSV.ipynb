# Titanic Survival Prediction 

## Data Dictionary
Here we got the data dictionary. This useful info will help you to understand the csv.

|Variable	|Definition	|Key|
|:----|:-----|:----|
|survival|	Survival|	0 = No, 1 = Yes|
|pclass|	Ticket class|	1 = 1st, 2 = 2nd, 3 = 3rd|
|sex|	Sex	| |
|Age|	Age in years| |	
|sibsp|	# of siblings / spouses aboard the Titanic| |
|parch|	# of parents / children aboard the Titanic|	 |
|ticket|	Ticket number|	 |
|fare|	Passenger fare|	 |
|cabin|	Cabin number| |	
|embarked|	Port of Embarkation|	C = Cherbourg, Q = Queenstown, S = Southampton |

## Variable Notes

**pclass:** A proxy for socio-economic status (SES)
1st = Upper
2nd = Middle
3rd = Lower

**age:** Age is fractional if less than 1. If the age is estimated, is it in the form of xx.5

**sibsp:** The dataset defines family relations in this way...
Sibling = brother, sister, stepbrother, stepsister
Spouse = husband, wife (mistresses and fianc√©s were ignored)

**parch:** The dataset defines family relations in this way...
Parent = mother, father
Child = daughter, son, stepdaughter, stepson
Some children travelled only with a nanny, therefore parch=0 for them.

Please check https://www.kaggle.com/c/titanic/data

# Drop some unnecessary columns

```python
df2=df.drop(labels=['Name','PassengerId','Cabin','Ticket'],axis=1)
df2.head()
```

# From column 'Sex', create another numeric column (sexc)
Randonforest cannot process text data. All data must be numeric type.

```python
code={'female':0,'male':1}
for k,i in df2.iterrows():
    df2.loc[k,"sexc"]=code[i['Sex']]
```

```python
df2.head()
```

# Count number of records of each category

```python
df2["Embarked"].value_counts()
```

# From column 'Embarked', create another numeric column (embarkedc)

```python

code={'C':0,'Q':1, 'S':2}
for k,i in df2.iterrows():
    if i['Embarked']==i['Embarked']:
        df2.loc[k,"embarkedc"]=code[i['Embarked']]
```

```python
df2.head(10)
```

```python
#remove the original column
df3=df2.drop(['Sex','Embarked'],1)
```

# Drop every NaN row

```python
df4=df3.dropna(axis=0)
df4.head(10)
```

# Alternative approuch--drop every NaN row using Masking

```python
df5=df3[ df3['Age'].notnull() ]
df5.head()
```

# Convert Dataframe to numpy array for traning

```python
df4.head()
```

```python
X=df4[df4.columns[1:]].values
x_names=list(df4.columns[1:])
```

```python
y=df4.Survived.values
y_names=['No','Yes']
```

```python
y.shape
```

```python
y=y.reshape(-1)
```

```python
y.shape
```

```python
X.shape
```

```python
X
```

# Construct Classifier

```python
from sklearn.ensemble import ExtraTreesClassifier,RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import cross_val_score
```

```python
forest = ExtraTreesClassifier(n_estimators=250,random_state=0)
```

# Lazy k-fold cross-valivation

```python
scores = cross_val_score(forest, X, y, cv=5)
scores.mean()  
```

# Training

```python
forest.fit(X,y)
```

# Training

```python
from sklearn.tree import export_graphviz
#forest = RandomForestClassifier(n_estimators=10,random_state=0, max_depth=3)
#forest.fit(X,y)
#estimator = forest.estimators_[0]

estimator=DecisionTreeClassifier(max_depth=2)
estimator.fit(X,y)
# Export as dot file
export_graphviz(estimator, out_file='tree.dot', 
                feature_names =  x_names,
                class_names = y_names,
                rounded = True, proportion = True, 
                precision = 1, filled = True)

# Convert to png using system command (requires Graphviz)
from subprocess import call
call(['dot', '-Tpng', 'tree.dot', '-o', 'tree.png', '-Gdpi=600'])

# Display in jupyter notebook
from IPython.display import Image
Image(filename = 'tree.png')
```

# Prediction

```python
#Pclass=3 (lower), Age=22 , SibSp=1, Parch=0, Fare=7.25 (free trip)
#sexc=1 (male), embarkedc=1 (Queenstown)
forest.predict([[3,22.0,1,0,7.25,1.0,1.0]])
```

```python
#Pclass=3 (lower), Age=22 , SibSp=1, Parch=0, Fare=7.25 (free trip)
#sexc=0 (male), embarkedc=1 (Queenstown)
forest.predict([[3,22.0,1,0,7.25,0.0,1.0]])
```
